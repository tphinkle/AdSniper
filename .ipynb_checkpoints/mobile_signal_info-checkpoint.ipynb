```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load the data

```python
directory='./data/'
sep_filename='mobile_signal_info_all_september.csv'
oct_filename='mobile_signal_info_all_october.csv'
nov_filename='mobile_signal_info_all_november.csv'

df_sep=pd.read_csv(directory+sep_filename)
df_oct=pd.read_csv(directory+oct_filename)
df_nov=pd.read_csv(directory+nov_filename)
```

```python
frames=[df_sep,df_oct,df_nov]
df_raw=pd.concat(frames)
```

# Map the data

```python
feature='cdma_dbm'
df=df_raw[(df_raw['longitude'].isnull()==False) & (df_raw['latitude'].isnull()==False) & (df_raw[feature].isnull()==False)]\
[['longitude', 'latitude', feature]]

df.head()
```

```python
long_mean=df['longitude'].values.mean()
lat_mean=df['latitude'].values.mean()
feature_mean=df[feature].values.mean()

print long_mean
print lat_mean
print feature_mean
```

```python
df[feature]=df[feature].map(lambda x: x-feature_mean)

feature_std=df[feature].std()

df.head()
```

```python
nlong=64
nlat=64
min_long=df['longitude'].min()
max_long=df['longitude'].max()
min_lat=df['latitude'].min()
max_lat=df['latitude'].max()
longs=np.linspace(min_long, max_long, nlong)
lats=np.linspace(min_lat, max_lat, nlat)
```

```python
avgs=np.zeros((nlong-1, nlat-1))
count=np.zeros((nlong-1, nlat-1))

for i in range(1,nlong):
    if i % 10==0:
        print 'i=', i
    for j in range(1,nlat):
        count[i-1,j-1]=len(df[(df['longitude'] > longs[i-1]) & (df['longitude'] < longs[i])\
                 & (df['latitude'] > lats[j-1]) & (df['latitude'] < lats[j])][feature])+1
        avgs[i-1,j-1]=df[(df['longitude'] > longs[i-1]) & (df['longitude'] < longs[i])\
                 & (df['latitude'] > lats[j-1]) & (df['latitude'] < lats[j])][feature].mean()/count[i-1,j-1]
```

```python
fig=plt.figure()
plt.imshow(avgs, vmin=feature_mean-1*feature_std, vmax=feature_mean+1*feature_std, origin='lower')
plt.colorbar()
plt.show()

fig.set_size_inches(10,10)
```

```python
count_mean=np.mean(count.flatten())
count_std=np.std(count.flatten())
print count.max()
print count_mean
print count_std

fig=plt.figure()
plt.imshow(count, vmin=0, vmax=count_mean+0.5*count_std, origin='lower')
plt.colorbar()
plt.show()

fig.set_size_inches(10,10)
```
